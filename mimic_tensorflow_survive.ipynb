#### values to floats, round by 1

```python
data = data.astype('float')
data = data.round(decimals=2)
data = data.fillna(0)
data
```

#### Data preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
survive = data.drop(columns='ID').copy()
survive.head()
```

```python
# features:
X = survive.drop(columns='survive').copy()

# make Normalize X:
#names = X.columns
#scaler = MinMaxScaler()
#X_rescaled = scaler.fit_transform(X)
#X = pd.DataFrame(data=X_rescaled, columns=names)

# target:
y = survive.survive
```

```python
# All columns are - numerical features.
num_features = X.columns
```

```python
# Make the pipeline to: 1) fill NaN values and 2) Use Standart Scale
transformer = make_pipeline( # Use sklearn.pipeline - make_pipeline.
        SimpleImputer(strategy='constant'), # possible: 'most_frequents' and 'mean'.
        StandardScaler(), # Standartize our data.
)

transformer
```

```python
preprocessor = make_column_transformer(
    (transformer, num_features)
)
preprocessor
```

```python
# Use train_test_split to split our data to taining and validation.
# stratify - make sure classes are evenlly represented across splits
X_train, X_valid, y_train, y_valid = \
    train_test_split(X, y, stratify=y, train_size=0.75)
```

```python
X_train
```

```python
X_train = preprocessor.fit_transform(X_train)
X_valid = preprocessor.transform(X_valid)
```

#### Optimum layers shape

```python
# #. Libraries we didn't have. (pip install scikeras. AND pip uninstall tensorflow-metal 
# #                                To success from scikeras.wrappers import KerasClassifier. on MacOS Apple Silicon )
import math

# from scikeras.wrappers import KerasClassifier

# from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# Difine costom function to find the layers numbers:
# Input: N_layers - hidden, fist_layer_nodes - input shape[1] (47),  last - output (1)
def FindLayerNodesLinear(n_layers, first_layer_nodes, last_layer_nodes):
    layers = []
    
    nodes_increment = (last_layer_nodes - first_layer_nodes)/ (n_layers-1)
    nodes = first_layer_nodes
    for i in range(1, n_layers+1):
        layers.append(math.ceil(nodes))
        nodes = nodes + nodes_increment
    
    return layers
```

```python
# In output: numbers of shape hidden layers (2 layers, 32, 17)
FindLayerNodesLinear(4, X.shape[1], 1)
```

#### Define the model

```python
import keras
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.metrics import AUC, Precision, Recall, F1Score
from tensorflow.keras.losses import MeanSquaredError as MSE
```

```python
model = keras.Sequential([ 
    
    layers.BatchNormalization(),
    layers.Dense(units=FindLayerNodesLinear(4, X_train.shape[1], 1)[0], input_dim=47 ,activation='relu'),
    layers.Dropout(rate=0.2),
    
    layers.BatchNormalization(),
    layers.Dense(units=FindLayerNodesLinear(4, X_train.shape[1], 1)[1], activation='relu'),
    layers.Dropout(rate=0.2),
    
        
    layers.BatchNormalization(),
    layers.Dense(units=FindLayerNodesLinear(4, X_train.shape[1], 1)[2], activation='relu'),
    layers.Dropout(rate=0.2),

    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy']
)
```

```python
early_stopping = keras.callbacks.EarlyStopping(
    patience=5,
    min_delta=0.01,
    restore_best_weights=True,
)
history = model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    shuffle=True,
    batch_size=60,
    epochs=50,
    callbacks=[early_stopping],
    verbose=0
)

history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot(title="binary_crossentropy")
#history_df.loc[:, ['accuracy', 'val_accuracy']].plot(title="accuracy")
history_df.loc[:, ['binary_accuracy', 'val_binary_accuracy']].plot(title="binary_accuracy")
```
